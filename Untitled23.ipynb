```python
import os
import pandas as pd
import numpy as np
import requests
import shutil
from tqdm import tqdm
import glob
import ast
import warnings

warnings.filterwarnings('ignore')
```

```python
os.makedirs('images', exist_ok=True)
os.makedirs('videos', exist_ok=True)
```

```python
csv_files = glob.glob('usc-x-24-us-election/part_*/*.csv.gz')

# Hey Ashiwin adjust the number of files to process
csv_files = csv_files[:1]
```

```python
def extract_media_urls(media_str):
    image_urls = set()
    video_urls = set()
    try:
        media_content = ast.literal_eval(media_str)
        if isinstance(media_content, list):
            for media_item in media_content:
                media_type = media_item.get('type', '')
                if media_type == 'photo':
                    media_url = media_item.get('media_url_https', None)
                    if media_url:
                        image_urls.add(media_url)
                elif media_type in ['video', 'animated_gif']:
                    video_info = media_item.get('video_info', {})
                    variants = video_info.get('variants', [])
                    max_bitrate = -1
                    best_variant_url = None
                    for variant in variants:
                        if 'bitrate' in variant:
                            if variant['bitrate'] > max_bitrate:
                                max_bitrate = variant['bitrate']
                                best_variant_url = variant['url']
                        else:
                            if not best_variant_url:
                                best_variant_url = variant['url']
                    if best_variant_url:
                        video_urls.add(best_variant_url)
    except Exception as e:
        print(f"Error parsing media column: {e}")
    return image_urls, video_urls
```

```python
def download_media(url, save_dir):
    try:
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)',
        }
        print(f"Attempting to download: {url}")
        if url.endswith('.m3u8'):
            filename = os.path.join(save_dir, f"{url.split('/')[-3]}.mp4")
            cmd = f'ffmpeg -y -loglevel error -headers "User-Agent: {headers["User-Agent"]}" -i "{url}" -c copy "{filename}"'
            print(f"Executing command: {cmd}")
            os.system(cmd)
            if os.path.exists(filename):
                print(f"Saved video: {filename}")
                return True
            else:
                print(f"Failed to save video from HLS stream: {url}")
                return False
        else:
            response = requests.get(url, headers=headers, stream=True, timeout=10)
            if response.status_code == 200:
                content_type = response.headers.get('content-type', '')
                print(f"Content-Type: {content_type}")
                if 'image' in content_type or any(ext in url.lower() for ext in ['.jpg', '.jpeg', '.png', '.gif']):
                    filename = os.path.join(save_dir, url.split('/')[-1].split('?')[0])
                    with open(filename, 'wb') as out_file:
                        shutil.copyfileobj(response.raw, out_file)
                    print(f"Saved image: {filename}")
                    return True
                elif 'video' in content_type or any(ext in url.lower() for ext in ['.mp4', '.avi', '.mov', '.wmv']):
                    filename = os.path.join(save_dir, url.split('/')[-1].split('?')[0])
                    with open(filename, 'wb') as out_file:
                        shutil.copyfileobj(response.raw, out_file)
                    print(f"Saved video: {filename}")
                    return True
                else:
                    print(f"Unrecognized content type for URL: {url}")
            else:
                print(f"Failed to download {url} (Status code: {response.status_code})")
    except Exception as e:
        print(f"Error downloading {url}: {e}")
    return False
```

```python
total_download_count = 0

for csv_file in csv_files:
    print(f"\nProcessing {csv_file}")
    try:
        df = pd.read_csv(csv_file, compression='gzip', low_memory=False)
    except Exception as e:
        print(f"Error reading {csv_file}: {e}")
        continue

    image_urls = set()
    video_urls = set()

    if 'media' in df.columns:
        media_items = df['media'].dropna().unique()
        for item in media_items:
            imgs, vids = extract_media_urls(item)
            image_urls.update(imgs)
            video_urls.update(vids)

    all_urls = list(image_urls.union(video_urls))

    print(f"Found {len(image_urls)} unique image URLs in {csv_file}")
    print(f"Found {len(video_urls)} unique video URLs in {csv_file}")
    print(f"Total unique media URLs: {len(all_urls)}")

    print(f"Processing {len(all_urls)} URLs (limited for testing)")

    print("Sample URLs to be processed:")
    for i, url in enumerate(all_urls):
        print(url)
        if i >= 9:
            break

    download_count = 0

    for url in tqdm(all_urls):
        if url:
            if any(ext in url.lower() for ext in ['.jpg', '.jpeg', '.png', '.gif']):
                if download_media(url, 'images'):
                    download_count += 1
            elif any(ext in url.lower() for ext in ['.mp4', '.avi', '.mov', '.wmv', '.m3u8']):
                if download_media(url, 'videos'):
                    download_count += 1
            else:
                try:
                    headers = {
                        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)',
                    }
                    response = requests.head(url, headers=headers, allow_redirects=True, timeout=5)
                    content_type = response.headers.get('content-type', '')
                    print(f"Content-Type from HEAD request: {content_type}")
                    if 'image' in content_type:
                        if download_media(url, 'images'):
                            download_count += 1
                    elif 'video' in content_type:
                        if download_media(url, 'videos'):
                            download_count += 1
                    else:
                        print(f"Content type not recognized: {content_type} for URL: {url}")
                except Exception as e:
                    print(f"Error getting content-type for {url}: {e}")

    print(f"Total files downloaded from {csv_file}: {download_count}")
    total_download_count += download_count

print(f"\nTotal files downloaded from all CSV files: {total_download_count}")
```

```python
import zipfile
import os
from google.colab import files

images_dir = 'images'
videos_dir = 'videos'
output_zip_file = 'media_content.zip'

with zipfile.ZipFile(output_zip_file, 'w') as zipf:
    for root, _, files in os.walk(images_dir):
        for file in files:
            file_path = os.path.join(root, file)
            zipf.write(file_path, os.path.relpath(file_path, images_dir))

    for root, _, files in os.walk(videos_dir):
        for file in files:
            file_path = os.path.join(root, file)
            zipf.write(file_path, os.path.relpath(file_path, videos_dir))

print(f"Compressed media files into {output_zip_file}")

files.download(output_zip_file)
```
